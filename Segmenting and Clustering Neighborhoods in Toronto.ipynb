## 1) Preprocessing and cleaning data

```python
df = df.rename(columns = {"Postal Code":"PostalCode"})
df.head()
```

```python
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['PostalCode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

## 2 Latitude and the Longitude Coordinates of each Neighborhood. 

Importing the Latitude and Longtitude CSV file

```python
lat_long = pd.read_csv("http://cocl.us/Geospatial_data")
lat_long.head()
```

```python
# Renaming the csv postal code to match the postalcode column of the main dataframe 
lat_long = lat_long.rename(columns = {"Postal Code":"PostalCode"})
lat_long.head()
```

We merge both dataframe together

```python
df_merge = pd.merge(df2, lat_long, on='PostalCode')
df_merge
```

## 3 Now working with rows which contains "Toronto" in their Borough.

```python
df_toronto = df_merge[df_merge['Borough'].str.contains("Toronto")].reset_index(drop=True)
df_toronto
```

### Plotting all the Neighbourhoods of the above data frame using Folium

```python
address = "Toronto, ON"

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Borough'], df_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
